```python
# Importar librerias
import matplotlib.pyplot as plt
import seaborn as sns

# Datos que importar o trabajar

categorias = ['A', 'B', 'C', 'D', 'E']
valores = [5, 8, 3, 9, 2]

# Figura de grafica de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Legends 
ax.set_xlabel(
    'Categrodias',
    fontsize=16,
    labelpad=10
)
ax.set_ylabel(
    'Valores',
    fontsize=16,
    labelpad=10
)

#Titulo
ax.set_title('Valores por categorias', fontsize=16, pad=15, fontweight='bold')


# Grafica de barras
sns.barplot(
    x= categorias, 
    y= valores, 
    ax=ax,

)
plt.show()
```
