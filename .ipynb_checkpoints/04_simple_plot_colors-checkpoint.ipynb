```python
# Importar librerias
import matplotlib.pyplot as plt
import seaborn as sns

# Datos que importar o trabajar

categorias = ['A', 'B', 'C', 'D', 'E'] 
valores = [5, 8, 3, 9, 2]

# Figura de grafica de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Color
sns.barplot(
    x= categorias, 
    y= valores, 
    ax=ax,
    # Esto cambiamos (customisable)
    palette=sns.color_palette("pastel", n_colors=len(categorias)), 
    hue=categorias, 
    legend=False
)
plt.show()
```
